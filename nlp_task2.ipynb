```python
import fasttext
import numpy as np
from huggingface_hub import hf_hub_download
# Load FastText model (you will need to download a pre-trained model or train your own)
# For demonstration purposes, let's assume you have a FastText model saved as "model.bin"
model_path = hf_hub_download(repo_id="facebook/fasttext-language-identification", filename="model.bin")
model = fasttext.load_model(model_path)

# User input for the keyword
user_input = input("Enter a keyword: ")

# List of words to compare against
word_list = list(data.keys())

most_similar_word = None
highest_similarity = -1  # Initialize with a low value
user_vector = model.get_word_vector(user_input)
for word in word_list:
    # Get the word vector for the current word in the list
    word_vector = model.get_word_vector(word)

    # Calculate cosine similarity between the user input and the current word
    similarity = np.dot(user_vector, word_vector) / (np.linalg.norm(user_vector) * np.linalg.norm(word_vector))

    if similarity > highest_similarity:
        highest_similarity = similarity
        most_similar_word = word

if most_similar_word:
    print("The most similar word is:", most_similar_word)
    print("Similarity score:", highest_similarity)
else:
    print("No similar word found.")
```
