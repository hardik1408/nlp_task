```python
import spacy
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
!python -m spacy download en_core_web_md
nlp = spacy.load("en_core_web_md")

paragraphs = list(data.values())

for paragraph in paragraphs:
  para = ''.join(paragraph)
  paragraph_embeddings = [nlp(para).vector]


user_word = input("Enter a word: ")

user_word_embedding = nlp(user_word).vector

similarities = [cosine_similarity([user_word_embedding], [paragraph_embedding])[0][0] for paragraph_embedding in paragraph_embeddings]

most_similar_index = np.argmax(similarities)
most_similar_paragraph = paragraphs[most_similar_index]
print("Most similar paragraph:", most_similar_paragraph)
```

```python
val=list(data.values())

result = ''.join(val[most_similar_index])
print(result)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
model_name = "facebook/bart-large-cnn"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
# Tokenize and generate summary
input_ids = tokenizer.encode("summarize: " + result, return_tensors="pt", max_length=1024, truncation=True)
summary_ids = model.generate(input_ids, max_length=150, min_length=50, length_penalty=2.0, num_beams=4, early_stopping=True)

# Decode and print the summary
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
print(summary)
```
