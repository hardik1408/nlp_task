```python
import torch
import numpy as np
from transformers import GPT2Tokenizer, GPT2Model

# Load the GPT-2 model and tokenizer
model_name = "gpt2"  # You can choose from various GPT-2 model sizes (e.g., gpt2, gpt2-medium, gpt2-large)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2Model.from_pretrained(model_name)

# User input for the keyword
user_input = input("Enter a keyword: ")

# Tokenize the user input
input_ids = tokenizer.encode(user_input, return_tensors="pt")

# Generate embeddings for the user input
with torch.no_grad():
    output = model(input_ids)
    user_embedding = output.last_hidden_state.mean(dim=1).squeeze().cpu().numpy()  # Convert to NumPy array

# Calculate the similarity between user input and dictionary keys
similarities = {}
threshold = 0.2  # Adjust the threshold as needed
for key in data.keys():
    # Tokenize the key
    key_ids = tokenizer.encode(key, return_tensors="pt")

    # Generate embeddings for the key
    with torch.no_grad():
        output = model(key_ids)
        key_embedding = output.last_hidden_state.mean(dim=1).squeeze().cpu().numpy()  # Convert to NumPy array

    # Calculate cosine similarity between user input and the key
    similarity = np.dot(user_embedding, key_embedding) / (np.linalg.norm(user_embedding) * np.linalg.norm(key_embedding))

    if similarity > threshold:
        similarities[key] = similarity

# Find the most similar key
if similarities:
    most_similar_key = max(similarities, key=lambda k: similarities[k])
    print("The most similar key to your input is:", most_similar_key)
else:
    print("No similar key found.")
```
